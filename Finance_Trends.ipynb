### Predictive Modelling


```python
# Create the Binary Target Variable (Y)
# Assuming 1-7 ranking scale, higher number means higher preference.
df['Prefers_High_Risk'] = (df['Equity_Market'] > df['Fixed_Deposits']).astype(int)

# Select Features (X)
feature_cols = ['age', 'gender', 'Objective', 'Purpose', 'Duration', 'Source']
X = df[feature_cols]

# One-Hot Encode Categorical Features
X_encoded = pd.get_dummies(X, columns=['gender', 'Objective', 'Purpose', 'Duration', 'Source'], drop_first=True)
Y = df['Prefers_High_Risk']

print("Shape of Encoded Features (X):", X_encoded.shape)
print("Distribution of Target (Y):")
print(Y.value_counts(normalize=True))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(
    X_encoded, Y, test_size=0.3, random_state=42, stratify=Y
)

# Train the Model
model = LogisticRegression(solver='liblinear', random_state=42)
model.fit(X_train, Y_train)

# Predict and Evaluate
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)

print(f"\nModel Accuracy on Test Set: {accuracy:.4f}")

# Get Feature Importance (Coefficients)
coefficients = pd.Series(model.coef_[0], index=X_encoded.columns)
# Sort by the absolute value to see the strongest predictors
top_predictors = coefficients.abs().sort_values(ascending=False).head(10)

print("\nTop 10 Predictors of Preferring High Risk:")
print(top_predictors)
```

### Final Conclusion: Driving Investment Strategy with Behavioral Data

The analysis of the 12,000-record Finance Trends Dataset goes beyond surface-level demographics to uncover the behavioral drivers that shape investment decisions.

1. Central Finding: Goals Outweigh Age

The predictive model (73.5% accuracy) reveals that an investor’s goals are the most powerful factor in deciding their risk level, much more so than their age. 

- Goals Control Risk Dominance:
The top predictor of high-risk appetite is the objective of Income (Coefficient ≈ 6.58), followed by Wealth Creation (Coefficient ≈ 3.14).
This shows that what an investor wants to achieve dictates their willingness to take risks more than who they are.

- Age Lacks Predictive Power:
Age and investment duration were found to be the least influential factors, confirming that within this young cohort, risk appetite is driven more by objectives than by age.

2. Gender-Based Risk Behavior

The analysis confirmed a distinct gender gap in risk tolerance.

- Female investors preferred low-risk assets like Fixed Deposits.

- Being male was a significant positive predictor of high-risk preference (Coefficient ≈ 1.18), confirming that gender is a necessaryinput in accurate risk profiling.

- Financial communication and product recommendations should therefore be gender-sensitive, reflecting these established behavioral differences.


3. Strategic Recommendations

These findings highlight two key strategies for financial firms:

- Adopt a Goal-First Risk Profile:
Replace age-based assessments with goal-oriented models prioritizing investment objectives (Income vs. Growth). This approach more accurately reflects a client’s risk appetite.

- Optimize Communication Channels:
Focus marketing for high-risk, high-return products on Television and Internet platforms, which showed the strongest predictive power (Coefficients ≈ 1.96 and 0.69 respectively) for influencing high-risk preference. This targets the people who are most interested.
